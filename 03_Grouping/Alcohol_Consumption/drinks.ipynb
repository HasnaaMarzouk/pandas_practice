### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

```python
data = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv")
```

### Step 3. Assign it to a variable called drinks.

```python
drinks = data
```

```python
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby("continent")["beer_servings"].mean().sort_values(ascending = False).head(1)
```

```python
drinks.groupby("continent")["beer_servings"].mean().sort_values(ascending = False).head(1).index[0]
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby("continent")["wine_servings"].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby("continent")[["total_litres_of_pure_alcohol","beer_servings","spirit_servings","wine_servings"]].mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby("continent")[["total_litres_of_pure_alcohol","beer_servings","spirit_servings","wine_servings"]].median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks[['spirit_servings']].agg(['mean', 'min', 'max'])
```

```python

drinks.groupby("continent")["spirit_servings"].agg(['mean','min','max'])
```
