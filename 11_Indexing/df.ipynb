### Step 3. Assign it to a variable called

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep = '\t')
df
```

#### 1. How to select the 'item_name' column?

```python
item_name_column = df['item_name']
item_name_column.head()
```

#### 2. How to select the 'item_name' and 'quantity' columns?

```python
item_name_quantity_columns = df[['item_name', 'quantity']]
item_name_quantity_columns.head()
```

#### 3. How to select the first 10 rows of the DataFrame?

```python
first_10_rows = df.head(10)
first_10_rows
```

#### 4. How to select rows where 'quantity' is greater than 1?

```python
quantity_greater_than_1 = df[df['quantity'] > 1]
quantity_greater_than_1.head()
```

#### 5. How to select the 'item_price' for the item at index 5 using `.loc` and `.iloc`?

```python
item_price_at_index_5_loc = df.loc[5, 'item_price']
print(f"Using .loc: {item_price_at_index_5_loc}")

item_price_at_index_5_iloc = df.iloc[5, df.columns.get_loc('item_price')]
print(f"Using .iloc: {item_price_at_index_5_iloc}")
```

#### 6. How to select the 'item_name' and 'item_price' columns for rows where 'quantity' is 2?

```python
item_name_price_quantity_2 = df[df['quantity'] == 2][['item_name', 'item_price']]
item_name_price_quantity_2.head()
```

#### 7. How to select all rows where 'item_name' is 'Chicken Bowl' and 'quantity' is 1?

```python
chicken_bowl_quantity_1 = df[(df['item_name'] == 'Chicken Bowl') & (df['quantity'] == 1)]
chicken_bowl_quantity_1.head()
```

#### 8. How to select rows from index 10 to 15 (inclusive) for all columns using `.iloc`?

```python
rows_10_to_15 = df.iloc[10:16]
rows_10_to_15
```

#### 9. How to select the 'item_price' for the last 3 rows using `.tail()`?

```python
item_price_last_3 = df['item_price'].tail(3)
item_price_last_3
```

#### 10. How to check if any 'item_name' contains the word 'Salad'?

```python
contains_salad = df['item_name'].str.contains('Salad').any()
contains_salad
```
